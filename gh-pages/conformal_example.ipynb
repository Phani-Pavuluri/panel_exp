
# Synthetic Control Method with Conformal Inference

This is a simple example of how to use the Synthetic Control Method to estimate the causal effect using the well known Kansas dataset. Using conformal inference to construct the confidence intervals.


```python
from panel_exp.panel_data import long_df_to_paneldataset
from panel_exp.methods.scm import SyntheticControl
import pandas as pd

long_df = pd.read_csv('kansas_parsed.csv')
panel_data = long_df_to_paneldataset(long_df, "year_qtr", "state", "lngdp", ["Kansas"], 2012)

print(panel_data.summarize())

```

```python
scm = SyntheticControl(inference='Conformal')
scm.run_analysis(panel_data)
```

```python
scm.plot()
```
