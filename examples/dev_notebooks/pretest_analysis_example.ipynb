```python
import sys
sys.path.insert(0, '../../')
from panel_exp.panel_data import long_df_to_paneldataset, PanelDataset, TimePeriod
```

```python
import pandas as pd 
import numpy as np

long_df = pd.read_csv('../data/kansas_parsed.csv', index_col=0)
long_df.iloc[:,0] = long_df.iloc[:,0].apply(lambda x: np.exp(x))
long_df.head()
```

```python
from panel_exp.design import CompleteRandomization, ThinningDesign, Rerandomization, QuickBlock, greedy_match_markets
from panel_exp.pretest_analysis.mde_ci_module import mde_ci_matrix
```

```python
panel_data = long_df_to_paneldataset(long_df, "year_qtr", "state", "lngdp")

gm = greedy_match_markets()
td = ThinningDesign(treatment_probability=0.5)
cr = CompleteRandomization(treatment_probability=0.5)
# rr= Rerandomization(treatment_probability=0.5, base_randomizer_cls=greedy_match_markets())

design = [gm,td,cr][0]

test_whitelist = ['Utah']

#to make test whitelist strictly inclusive, you must pass test_blacklist in the below manner (i.e. include all dmas not part of test_whitelist)
# test_blacklist = [geo for geo in panel_data.units.tolist() if geo not in test_whitelist]

gm_panel_data = design.assign(panel_data=panel_data,test_whitelist=test_whitelist)
```

```python
print('Test group: %s \n\nControl group: %s' %(gm_panel_data.treated_units, [unit for unit in gm_panel_data.units if unit not in gm_panel_data.treated_units]))
```

Below is the sensitivity analysis for test and control groups identified by the selected design. Additionally, we also provide sensitivity analysis for trimmed versions of test group incase we need to reduce the test size due to spend constraints.

```python
results = mde_ci_matrix(gm_panel_data)
```

```python
#Minimum detectable effect for varying test sizes
results['mde_scores']
```

```python
#Confidence Interval for varying test sizes
results['ci_scores']
```

Trim control group for a given test group - if we were to choose trimmed version of test group from the results above for spend reasons, then we want to ensure control group is also updated accordingly - i.e. check if we can trim down control without compromising on sensitivity. This way we can run the test with smaller control and save those trimmed geos for another test. 

As an example, I'm picking trimmed test group from row 6 where 8 week MDE and CI are 0.41%, 48.9K respectively, then apply control trim to see if it's possible to further improve sensitivity by trimming down the control group. 

```python
#Limiting the raw data to just control and test geos
trimmed_long_df = long_df[long_df.state.isin(results['ci_scores'].iloc[6,7] + results['ci_scores'].iloc[6,8])]
```

```python
#setting treated units
treated_units = results['ci_scores'].iloc[6,7]
```

```python
#creating panel dataset with treated units
trimmed_panel_data = long_df_to_paneldataset(trimmed_long_df, "year_qtr", "state", "lngdp", treated_units = treated_units)
```

```python
trimmed_results = mde_ci_matrix(trimmed_panel_data, test_trim='no', control_trim='yes')
```

```python
trimmed_results['mde_scores']
```

```python
trimmed_results['ci_scores']
```

As you can see from row 4 in the above results that 8 week MDE and CI (0.32%, 38K) significantly improved after trimming down control from 15 geos to 11 geos.

To recap, if we are using non-trimmed version of the test group then no further trimming of control is needed but if we decide to use trimmed version of test group due to spend constraints then we'd want to check if applying control trim yeilds better sensitivity.
